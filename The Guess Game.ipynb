```python
import random
name = input("Enter you name: ")
print("Welcome " + name + " to the Guess Game")
print('''Rules: 1. The task is to know the secret word in the game
2. Enter the Character if the letter is in the word and you guess the word right you win
3. you have limited chances to attemp

Are You Ready to Guess the secret word!

Lets Get Started''')
words = ["Magic","Greeting","World","Positive","Science","Learning","Artificial"]
word = random.choice(words)
word_lo = word.lower()
print("Guess the Character in the word")
guesses = ''

Turns = 15
while True:
    fail = 0
    for char in word_lo:
        if char in guesses:
            print(char)
        else:
            print("*")
            fail += 1
            
    if fail == 0:
        print("You Win")
        print("The word is: ", word_lo)
        break
        
    guess = input("Guess a charater: ")
    
    guesses += guess
    
    if guess not in word_lo:
        Turns -= 1
        print("wrong try again")
        print("you have "+ str(Turns) + " more guess left")
            
    if Turns == 0:
        print("You loose")
        print("The word is: ", word_lo)
        break
```
